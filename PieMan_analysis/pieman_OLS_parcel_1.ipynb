```python
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# ------------------------------------------------------------------
# 1) FILE PATHS  (edit to match your drive)
# ------------------------------------------------------------------
neural_path = "/Volumes/Passport/fmriprep/derivatives/RDMs/sub-002_run-1_pieman_parcel-1_RDM.npy"
behav_path  = "/Volumes/Passport/fmriprep/derivatives/RDMs_behavior/1Pieman_affectionate_RDM.npy"
```

```python


# ------------------------------------------------------------------
# 2) LOAD RDMs
# ------------------------------------------------------------------
RDM_neural = np.load(neural_path)   # shape (160,160)
RDM_behav  = np.load(behav_path)    # shape (160,160)

# ------------------------------------------------------------------
# 3) EXTRACT ONE TRIANGLE (lower inc. diagonal)
# ------------------------------------------------------------------
i_low = np.tril_indices(160, k=-1)   # change k=-1 to drop diag if needed

y = RDM_neural[i_low]               # dependent variable (target)
affectionate = RDM_behav[i_low][:, None]       # predictor matrix (N×p) -> shape (12880,1)
affectionate = sm.add_constant(affectionate)                # add intercept (constant) to predictor matrix

# ------------------------------------------------------------------
# 4) OLS FIT
# ------------------------------------------------------------------

model = sm.OLS(y, affectionate).fit()

model.summary()

# beta_x1   = model.params['x1']

# beta = ols.coef_[0]
# r2   = ols.score(X, y)

# print("RSA (OLS) results for sub‑002, 1Pieman, parcel‑1")
# print(f"β_affectionate = {beta:.4f}")
# print(f"R²             = {r2:.4f}")
```
